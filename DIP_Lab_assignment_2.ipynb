```python
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
```

```python
image_url = "https://images.unsplash.com/photo-1516298252535-cf2ac5147f9b?fm=jpg&q=60&w=3000&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxzZWFyY2h8M3x8c3BvcnRzJTIwY2FyfGVufDB8fDB8fHww"
```

```python
# Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
```

```python
# Decode the image
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
```

```python
# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#show real image
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Real Image")
plt.show()
```

```python
#Resized image
resized_image = cv2.resize(img_rgb, (256, 256))

plt.imshow(resized_image)
plt.axis('off')
plt.title("Resized Image")
plt.show()
```

```python
#Rotated images
rotated_image = cv2.rotate(resized_image, cv2.ROTATE_180)

plt.imshow(rotated_image)
plt.axis('off')
plt.title("Resized Image")
plt.show()
```

```python
#Vertically_flipped_image
vertical_flipped_image = cv2.flip(resized_image, 0)


plt.imshow(vertical_flipped_image)
plt.axis('off')
plt.title("Vertical Flipped Image")
plt.show()
```
