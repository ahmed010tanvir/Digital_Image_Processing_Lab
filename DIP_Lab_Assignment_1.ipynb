**Task-1: Extract a 100Ã—100 Region from the Center of the Image read from an URL**

```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
image_url = "https://images.unsplash.com/photo-1599052114690-2db742187198?q=80&w=1770&auto=format&fit=crop&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"
```

```python
# Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
```

```python
# Decode the image
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
```

```python
# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.title("A sunrise image from URL")
plt.show()
```

```python
height, width = img_rgb.shape[:2]

# Extract 100x100 region from center
extracted_region = img_rgb[
   height//2 -50  : height//2 +50 ,
   width//2 -50 : width//2 +50
]


# Show extracted region
plt.imshow(extracted_region)
plt.axis("off")
plt.title("Extracted 100x100 Region")
plt.show()
```

**Task 2: Convert the image to HSV color space, visualize each components in grayscale representation**

```python
# Convert BGR to HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Split the HSV image into its components
hue, saturation, value = cv2.split(img_hsv)
```

```python
# create a figure  then display original image and HSV images
fig, axes = plt.subplots(1, 4, figsize=(20, 4))


axes[0].imshow(img_hsv)
axes[0].set_title("HSV Color Space")
axes[0].axis('off')


axes[1].imshow(hue, cmap='gray')
axes[1].set_title("Hue Channel")
axes[1].axis('off')


axes[2].imshow(saturation, cmap='gray')
axes[2].set_title("Saturation Channel")
axes[2].axis('off')


axes[3].imshow(value, cmap='gray')
axes[3].set_title("Value Channel")
axes[3].axis('off')

plt.tight_layout()
plt.show()
```
