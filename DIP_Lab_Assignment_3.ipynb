```python
import numpy as np
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img1 = np.ones((500, 500, 3), dtype=np.uint8) * 255
cv2.rectangle(img1,(100,150),(400,350),(0,255,0),-1) #BGR
cv2_imshow(img1)
```

```python
img2 = np.ones((500, 500, 3), dtype=np.uint8) * 255
cv2.circle(img2, (250, 250), 100, (255, 0, 255), -1)
cv2_imshow(img2)
```

**Bitwise AND**

```python
bitwise_and = cv2.bitwise_and(img1, img2)
cv2_imshow(bitwise_and)
```

**Bitwise OR**

```python
bitwise_or = cv2.bitwise_or(img1, img2)
cv2_imshow(bitwise_or)
```

**Bitwise XOR**

```python
bitwise_xor = cv2.bitwise_xor(img1, img2)
cv2_imshow(bitwise_xor)
```

**Bitwise NOT-Image 1**

```python
bitwise_not1 = cv2.bitwise_not(img1)
cv2_imshow(bitwise_not1)
```

**Bitwise NOT-Image 2**


```python
bitwise_not2 = cv2.bitwise_not(img2)
cv2_imshow(bitwise_not2)
```

**Element-wise Multiplication**

```python
multiplication = cv2.multiply(img1, img2)
cv2_imshow(multiplication)
```

**Alpha Bending**

```python
alpha_bending = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
cv2_imshow(alpha_bending)
```


**Image Subtraction-Difference between two images**

```python
subtraction = cv2.subtract(img1, img2)
cv2_imshow(subtraction)
```

```python
subtraction = cv2.subtract(img2, img1)
cv2_imshow(subtraction)
```
